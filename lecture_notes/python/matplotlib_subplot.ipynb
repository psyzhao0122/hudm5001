```python
import matplotlib.pyplot as plt # `plt` is an alias for the `matplotlib.pyplot` module.
import numpy as np
import pandas as pd
import seaborn as sns # (wrapper of matplotlib)
```

```python
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv'
df = pd.read_csv(url, sep=';') # load wine data
df.head() # view the first 5 rows of the dataframe
```

- Challenge: Given response variable y and a list of X variables, write a *loop* that builds graphs of y versus each x in X.  
  Hint: include `plt.show()` after each graph to show them all together.

**The original answer:**

A problem: `plt.show()` inside the loop is used. `plt.show()` inside the loop creates individual figures for each plot, which is not what we want when using subplots. Instead, we want to create the subplots in one figure and display them together.

```python
y = df.pH
X = ['chlorides','citric acid','density']

nrows = len(X)

for row in range(nrows):
    plt.subplot(nrows, 1, row+1)
    plt.scatter(df[X[row]], y)
    plt.title('pH vs {}'.format(X[row]))
    plt.xlabel(X[row]) 
    plt.ylabel('pH') 
    plt.show()
```

**The modified code:**

```python
y = df.pH
X = ['chlorides','citric acid','density']

nrows = len(X)

for row in range(nrows):
    plt.subplot(nrows, 1, row+1)
    plt.scatter(df[X[row]], y)
    plt.title('pH vs {}'.format(X[row]))
    plt.xlabel(X[row]) 
    plt.ylabel('pH') 

plt.tight_layout()  # adjust spacing between subplots
plt.show()
```

**Replace `row+1` in `plot.subplot()` with `1` and observe the result.**

```python
y = df.pH
X = ['chlorides','citric acid','density']

nrows = len(X)

for row in range(nrows):
    plt.subplot(nrows, 1, 1)
    plt.scatter(df[X[row]], y)
    plt.title('pH vs {}'.format(X[row]))
    plt.xlabel(X[row]) 
    plt.ylabel('pH') 

plt.tight_layout()  # adjust spacing between subplots
plt.show()
```
